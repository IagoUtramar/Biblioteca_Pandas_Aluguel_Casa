### Fazendo uma seleção por filtragem

dataset = dataset[(dataset["coluna"]==premissa)]

pode colocar expressões diferentes (<=,>=,>,<,==)

```python
aluguel_casa = aluguel_casa[(aluguel_casa["total"]<=4000)]
aluguel_casa.head()
```

```python
aluguel_casa.describe()
```

### Filtragem condicional

dataset = dataset[(dataset["coluna"]==premissa)] & [(dataset["coluna"]==premissa)]

mesclando com função "ou" é caracterizada por essa barra em pé |


dataset = dataset[(dataset["coluna"]==premissa)] & [(dataset["coluna"]==premissa)& ((dataset["coluna"]==premissa)|(dataset["coluna"]==premissa))]

```python
aluguel_casa = aluguel_casa[(aluguel_casa["total"]<=4000) & (aluguel_casa['animal']=='accept') & ((aluguel_casa['rooms']==2)|(aluguel_casa['rooms']==3))]
aluguel_casa.head()
```

### Condicional com .query
dataset = dataset.query("coluna<=premissa and coluna =='premissa' and (coluna == premissa or coluna== premissa)")

```python
aluguel_casa = aluguel_casa.query("total<=4000 and animal=='accept' and (rooms==2 or rooms==3)")
aluguel_casa.head()
```

```python
aluguel_casa.describe()
```

### Agrupamento de dados

o agrupamento de dados é exatamente igual a uma tabela dinâmica no excel.

agg é agrupamento

.groupby(['coluna principal'])['coluna de valor'].agg("mean")  #mean é media

.reset_index() #coloca em formato de data frame

```python
agg_city_price = aluguel_casa.groupby(['city'])['total'].agg("mean").reset_index()
agg_city_price.head()
```

```python
agg_bathroom = aluguel_casa.groupby(['bathroom','rooms','city'])['total'].agg("mean").reset_index()
agg_bathroom.head()
```
